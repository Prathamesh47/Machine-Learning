```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sms
```

```python
df=pd.read_csv('cities_r2.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
x=df[['total_graduates']]
print(x)
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel("wcss")
plt.title('Knee curve')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
df['clusters']=kmeans.fit_predict(x)
```

```python
print(df[['name_of_city','total_graduates','clusters']])
```

```python
sms.scatterplot(x='name_of_city',y='total_graduates',hue='clusters',data=df)
plt.xlabel('name of city')
plt.ylabel('total graduates')
plt.title('KMeans clustering')
plt.show()
```
