```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('madfhantr.csv')
df.head(5)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
df['Gender']=le.fit_transform(df['Gender'])
df['Married']=le.fit_transform(df['Married'])
df['Education']=le.fit_transform(df['Education'])
df['Self_Employed']=le.fit_transform(df['Self_Employed'])
df['Property_Area']=le.fit_transform(df['Property_Area'])
df['Loan_ID']=le.fit_transform(df['Loan_ID'])
df.head()
```

```python
df.dtypes
```

```python
dependent_mapping = {'0' : 0, '1' : 1, '2': 2, '3+': 3}
df['Dependents'] = df['Dependents'].map(dependent_mapping)

df
```

```python
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mean())
df.isnull().sum()
```

```python
x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
df.shape
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
reg=DecisionTreeClassifier()
reg.fit(xtrain,ytrain)
```

```python
ypred=reg.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
accuracy = accuracy_score(ytest, ypred)
print(f'Accuracy: {accuracy}')
```

```python
cunf = confusion_matrix(ytest, ypred)
print(cunf)
```
